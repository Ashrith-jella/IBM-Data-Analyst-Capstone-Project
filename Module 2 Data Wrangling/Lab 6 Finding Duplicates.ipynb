## Identify and Analyze Duplicates


### Task 1: Identify Duplicate Rows
1. Count the number of duplicate rows in the dataset.
3. Display the first few duplicate rows to understand their structure.


```python
## Write your code here
df.duplicated().sum()
# Display first few duplicate rows
df[df.duplicated()].head()
```

### Task 2: Analyze Characteristics of Duplicates
1. Identify duplicate rows based on selected columns such as MainBranch, Employment, and RemoteWork. Analyse which columns frequently contain identical values within these duplicate rows.
2. Analyse the characteristics of rows that are duplicates based on a subset of columns, such as MainBranch, Employment, and RemoteWork. Determine which columns frequently have identical values across these rows.
   


```python
## Write your code here
duplicate_subset = df[df.duplicated(subset=["MainBranch", "Employment", "RemoteWork"], keep=False)]
duplicate_subset.head()
grouped = duplicate_subset.groupby(["MainBranch", "Employment", "RemoteWork"])

identical_columns = {}

for col in df.columns:
    if col not in ["MainBranch", "Employment", "RemoteWork"]:
        identical_count = grouped[col].nunique().eq(1).sum()
        total_groups = grouped.ngroups
        identical_columns[col] = f"{(identical_count / total_groups) * 100:.2f}% identical across groups"

for col, stat in identical_columns.items():
    print(f"{col}: {stat}")
```

### Task 3: Visualize Duplicates Distribution
1. Create visualizations to show the distribution of duplicates across different categories.
2. Use bar charts or pie charts to represent the distribution of duplicates by Country and Employment.


```python
## Write your code here
!pip install seaborn
duplicates = df[df.duplicated(keep=False)]
country_counts = duplicates["Country"].value_counts()
employment_counts = duplicates["Employment"].value_counts()

import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12,6))
sns.barplot(x=country_counts.index[:10], y=country_counts.values[:10], palette='viridis')
plt.xticks(rotation = 45)
plt.title('Top 10 Countries with Most Duplicates')
plt.xlabel("Country")
plt.ylabel('Number of Duplicates')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 8))
duplicates['Employment'].value_counts().head(6).plot.pie(
    autopct='%1.1f%%',
    startangle=90,
    colors=sns.color_palette("pastel"),
    wedgeprops={'edgecolor': 'black'}
)
plt.title("Duplicate Distribution by Employment")
plt.ylabel("")  # Hide y-axis label
plt.show()
```

### Task 4: Strategic Removal of Duplicates
1. Decide which columns are critical for defining uniqueness in the dataset.
2. Remove duplicates based on a subset of columns if complete row duplication is not a good criterion.


```python
print(df.columns.tolist())
```

```python
## Write your code here
df_cleaned = df.drop_duplicates(subset=["MainBranch", "Employment", "RemoteWork", "Country", "Age"])
print("Original rows:", len(df))
print("After removal:", len(df_cleaned))
print("Duplicates removed:", len(df) - len(df_cleaned))

```

---

### **Duplicate Identification and Removal Process**

1. **Identifying Duplicates:**
   - We used `df.duplicated()` to find completely identical rows.
   - Then used `df.duplicated(subset=[...])` to identify partial duplicates based on key profile columns.

3. **Removing Duplicates:**
   - Used `df.drop_duplicates(subset=[...])` to remove repeated profiles while retaining diverse responses.
   - This retained one representative record for each unique combination of selected columns.

4. **Results:**
   - Original rows: **66,586**
   - After removal: **12,210**
   - Duplicates removed: **54,376**

5. **Reasoning:**
   - Complete row duplication was rare, so a subset-based approach preserved essential diversity while eliminating redundancy.
